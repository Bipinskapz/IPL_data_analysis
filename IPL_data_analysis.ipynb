# Match won with closest margin

```python
win_by_runs = data[data['win_by_runs'].ge(1)]
win_by_runs
```

```python
w = win_by_runs['win_by_runs']
w
```

```python
data.iloc[w.idxmin()]
```

# Which season had most no. of matches ???

# Most successful IPL team

```python
success_team = data.winner.value_counts()
success_team
```

```python
team = data.team1.unique()
team
```

```python
plt.rcParams['figure.figsize'] = (14,8)
sns.barplot(x=success_team.index,y=success_team,orient='v')
plt.show()
```

# Has Toss Winning Helped In Winning The Match ???

```python
toss_winning_match_winning = data['toss_winner'] == data['winner']
toss_winning_match_winning.groupby(toss_winning_match_winning).size()
```

```python
sns.countplot(x=toss_winning_match_winning)
```
